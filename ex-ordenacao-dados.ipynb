#### Bubble Sort

```python
def bubble_sort(lista):
    n = len(lista)

    for i in range(n):
        for j in range(n-i-1):
            if lista[j] > lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]
    return lista
```

```python
%timeit bubble_sort(numeros.copy())
```

```python
%timeit bubble_sort(numeros2.copy())
```

#### Selection Sort

```python
def select_sort(lista):
    n = len(lista)

    for i in range(n):
        min_index = i
        for j in range(i+1,n):
            if lista[j] < lista[min_index]:
                min_index = j
        lista[i], lista[min_index] = lista[min_index],lista[i]
    
    return lista
```

```python
%timeit select_sort(numeros.copy())
```

```python
%timeit select_sort(numeros2.copy())
```

#### Insertion Sort

```python
def insert_sort(lista):
    for i in range(1,len(lista)):
        chave = lista[i]
        j = i - 1

        while j >= 0 and lista[j] > chave:
            lista[j+1] = lista[j]
            j -= 1

        lista[j+1] = chave
    
    return lista
```

```python
%timeit insert_sort(numeros.copy())
```

```python
%timeit insert_sort(numeros2.copy())
```

#### Shell Sort

```python
def shell_sort(vetor):

    intervalo = len(vetor)//2

    while intervalo > 0:
        for i in range(intervalo,len(vetor)):
            temp = vetor[i]
            j=i
            
            while j >= intervalo and vetor[j - intervalo] > temp:
                vetor[j] = vetor[j - intervalo]
                j -= intervalo

            vetor[j] = temp

        intervalo //= 2

    return vetor
```

```python
%timeit shell_sort(numeros.copy())
```

```python
%timeit shell_sort(numeros2.copy())
```

#### Merge Sort

```python
def merge_sort(vetor):
    if len(vetor) > 1:
        divisao = len(vetor)//2
        esquerda = vetor[:divisao].copy()
        direita = vetor[divisao:].copy()

        merge_sort(esquerda)
        merge_sort(direita)

        i = j = k = 0

    #Ordena a esquerda e direita
        while i < len(esquerda) and j < len(direita):
            if esquerda[i] < direita[j]:
                vetor[k] = esquerda[i]
                i += 1
            else:
                vetor[k] = direita[j]
                j += 1

            k += 1
    #Ordenação final
        while i < len(esquerda):
            vetor[k] = esquerda[i]
            i += 1
            k += 1
        while j < len(direita):
            vetor[k] = direita[j]
            j += 1
            k += 1
        return vetor
```

```python
%timeit merge_sort(numeros.copy())
```

```python
%timeit merge_sort(numeros2.copy())
```

#### Quick Sort

```python
def particao(vetor, inicio, final):
    meio = (inicio + final) // 2
    vetor[meio], vetor[final] = vetor[final], vetor[meio]

    pivo = vetor[final]
    i = inicio - 1

    for j in range(inicio, final):
        if vetor[j] <= pivo:
            i += 1
            vetor[i], vetor[j] = vetor[j], vetor[i]

    vetor[i + 1], vetor[final] = vetor[final], vetor[i + 1]
    return i + 1


def quick_sort(vetor, inicio, final):
    if inicio < final:
        posicao = particao(vetor, inicio, final)
        quick_sort(vetor, inicio, posicao - 1)
        quick_sort(vetor, posicao + 1, final)
    return vetor
```

```python
%timeit quick_sort(numeros.copy(), 0, len(numeros)-1)
```

```python
%timeit quick_sort(numeros2.copy(), 0, len(numeros2.copy())-1)
```
