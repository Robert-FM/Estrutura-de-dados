```python
import csv

with open("dados.csv", "w", newline="") as f:
    escritor = csv.writer(f)
    escritor.writerow(["Nome", "Idade"])
    escritor.writerow(["Ana", 25])
```

```python
with open("dados.csv", "r") as f:
    leitor = csv.reader(f)
    for linha in leitor:
        print(linha)
```

```python
import json

# escrever
dados = {"nome": "Andrea", "idade": 30}

with open("dados.json", "w") as f:
    json.dump(dados, f)

# ler
with open("dados.json", "r") as f:
    dados = json.load(f)
    print(dados)
```
